### 2.2 Problem Solving

```python
#Dealing with missing numeric values
df.fillna(df.mean(numeric_only=True), inplace=True)
print (df.isnull().sum()) # Check for missing values
```

```python
#Drop duplicates
df=df.drop_duplicates()
print(df.duplicated().sum()) # Check for duplicates values
```

```python
# Convert social media time column from minutes to hours
df['social_media_time_min'] = df["social_media_time_min"] / 60
```

```python
# Convert screen time column from minutes to hours
df['daily_screen_time_min'] = df['daily_screen_time_min'] / 60
```

```python
#Rename columns
df.rename(columns={"daily_screen_time_hours":"Daily_Screen_Time_hours"},inplace=True)
df.rename(columns={"num_app_switches":"Num_App_Switches"},inplace=True)
df.rename(columns={"sleep_hours":"Sleep_Hours"},inplace=True)
df.rename(columns={"notification_count":"Notification_Count"},inplace=True)
df.rename(columns={"social_media_time_min":"Social_Media_Time"},inplace=True)
df.rename(columns={"focus_score":"Focus_Score"},inplace=True)
df.rename(columns={"mood_score":"Mood_Score"},inplace=True)
df.rename(columns={"anxiety_level":"Anxiety_Level"},inplace=True)
df.rename(columns={"digital_wellbeing_score":"Digital_Wellbeing_Score"},inplace=True)
df.rename(columns={"daily_screen_time_min":"Daily_Screen_Time"},inplace=True)

df.info()
```

```python
#Convert data types
df["Num_App_Switches"]=df["Num_App_Switches"].astype(int)
df["Notification_Count"]=df["Notification_Count"].astype(int)
df.info()
```

## 3. Exploratory Data Analysis (EDA)

###3.1 Statistical Analysis

```python
# Check Records, Data Types, Duplicated Values, Missing values.
print("Dataset shape (rows, columns):", df.shape) # Check for records
print("\nData types:\n", df.dtypes) # Check for data types
print("\nNumber of duplicated rows:", df.duplicated().sum()) # Check for duplicated values
print("\nMissing values per column:\n", df.isnull().sum()) # Check for missing values
```

```python
# Descriptive statistics
desc_stats = pd.DataFrame({
    'Mean': df.mean(),
    'Median': df.median(),
    'Std Dev': df.std(),
    'Min': df.min(),
    'Max': df.max()
})

# Round to 2 decimals
desc_stats_rounded = desc_stats.round(2)
desc_stats_rounded
```

```python
# Boxplots to visualize distributions and outliers
plt.figure(figsize=(18, 5))

plt.subplot(1, 5, 1)
sns.boxplot(y=df['Sleep_Hours'])
plt.title('Boxplot of Sleep Hours')

plt.subplot(1, 5, 2)
sns.boxplot(y=df['Daily_Screen_Time'] / 60)
plt.title('Boxplot of Daily Screen Time (Hr)')

plt.subplot(1, 5, 3)
sns.boxplot(y=df['Notification_Count'])
plt.title('Boxplot of Notifications')

plt.subplot(1, 5, 4)
sns.boxplot(y=df['Focus_Score'])
plt.title('Boxplot of Focus Score')

plt.subplot(1, 5, 5)
sns.boxplot(y=df['Anxiety_Level'])
plt.title('Boxplot of Mood Score')



plt.tight_layout()
plt.show()
```

```python
# Correlation matrix and heatmap
plt.figure(figsize=(10, 8))
corr = df.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f", square=True, linewidths=0.6)
plt.title('Correlation Heatmap')
plt.show()
```

```python
sns.scatterplot(x='Focus_Score', y='Daily_Screen_Time', data=df)
plt.title("Focus Score vs Social Media Time")
plt.show()
```

```python
numeric_cols = ['Daily_Screen_Time', 'Sleep_Hours', 'Social_Media_Time', 'Anxiety_Level', 'Digital_Wellbeing_Score']

plt.figure(figsize=(16, 10))

for i, col in enumerate(numeric_cols, 1):
    plt.subplot(2, 3, i)
    sns.histplot(df[col], kde=True, bins=30)
    plt.title(f'Distribution of {col}')

plt.tight_layout()
plt.show()
```

### 3.2 Visual Analysis

```python
# scatter plot of App Switches vs Focus Score
plt.figure(figsize=(10,5))
sns.regplot(data=df, x="Num_App_Switches" , y="Focus_Score",ci=None)
plt.title('relationship between app switching and focus')
plt.xlabel('Number of App Switches')
plt.ylabel('Focus Score')
plt.grid(True)
plt.show()
```

```python
# scatter plot of Daily Screen Time vs Sleep Hours
plt.figure(figsize=(10,5))
sns.regplot(data=df, x="Daily_Screen_Time" , y="Sleep_Hours",ci=None)
plt.title('relationship between Daily Screen Time and Sleep Hours')
plt.xlabel('Daily Screen Time')
plt.ylabel('Sleep Hours')
plt.grid(True)
plt.show()
```
